# Implementation of Leave-One-Out cross validation
## Step 1: Import libraries

```python
from sklearn.model_selection import train_test_split
# from sklearn.model_selection import LeaveOneOut
from sklearn.model_selection import LeaveOneOut
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
import numpy as np
import pandas as pd
```

## Step 2: Read dataset

```python
#importing the inbuilt dataset in sklearn package
from sklearn import datasets
diabetes_data = datasets.load_diabetes()
```

```python
print(diabetes_data.feature_names)
```

```python
print(diabetes_data.data)
```

```python
diabetes_df = pd.DataFrame(diabetes_data.data, columns = diabetes_data.feature_names)
diabetes_df['Progression'] = diabetes_data.target
diabetes_df
```

```python
X = diabetes_df.drop(columns='Progression',axis='columns')
Y = diabetes_df['Progression']
X
```

## Split dataset into training and test datasets

```python
# X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 10)
```

## Implement Leave-One-out Cross Validation 

```python
#define cross-validation method to use
l1o_cv = LeaveOneOut()

#build multiple linear regression model
model = LinearRegression()

#use Leave-P-Out cross validation to evaluate model
scores = cross_val_score(model, X, Y, scoring='neg_mean_absolute_error',
                         cv=l1o_cv, n_jobs=-1)
```

## Calculate mean absolute error

```python
#view mean absolute error
np.mean(np.absolute(scores))
```
