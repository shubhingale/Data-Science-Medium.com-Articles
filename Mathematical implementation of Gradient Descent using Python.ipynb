# Mathematical implementation of Gradient Descent using Python
### Following is the problem statement to implement gradient descent.
1. Cost function : $f(x) = x^3 -4x^2+6$
2. An equation of slope(Find derivative of cost function $f(x)$ w.r.t. $x$. $f'(x) = 3x^2-8x$
3. Updation Rule is as follows

<center><I>Next weight = Existing Weight - Learning Rate * Gradient</I>

    OR
$$x_{k+1}=x_k- \alpha f'(x_k)$$

Let us assume initial values of $x_k$ and  $\alpha$ as 0.5 and 0.05 respectively.

<b>Step 1:</b> Define $f'(x)$ and $x_{k+1}$ equations.

```python
f_x = lambda x: pow(x,3)-(4*pow(x,2))+6
fd_x = lambda x: (3*pow(x,2))-(8*x)
Calculate_X_next = lambda x_k:x_k - (learning_rate * fd_x(x_k))
```

<b>Step 2:</b> Iteratively, calculate $x_k$ and $f'(x)$, till it stops changing value of $x_k$.

```python
import pandas as pd
learning_rate = 0.05
x_k = 0.5
df = pd.DataFrame(columns={'X_k','fd_x'})
df = df.astype(float)
df = df.append({'X_k':x_k,'fd_x':fd_x(x_k)},ignore_index=True
)
for i in range(35):
    x_k = Calculate_X_next(x_k)
    df = df.append({'X_k':x_k,'fd_x':fd_x(x_k)},ignore_index=True
)
df
```

<b>Step 3:</b> Plot a plot $f'(x)$ vs $x_{k}$.

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()
mark = [df.fd_x.idxmin()]
ax.plot(df.X_k, df.fd_x)
ax.plot(df.X_k, df.fd_x,markevery=mark, ls="", marker="o", label="points")

def annot_min(x,y, ax=None):
    xmin = x[df.fd_x.idxmin()]
    ymin = y[df.fd_x.idxmin()]
    text= "x={:.3f}, y={:.3f}".format(xmin, ymin)
    if not ax:
        ax=plt.gca()
    bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
    arrowprops=dict(arrowstyle="->",connectionstyle="angle,angleA=0,angleB=90")
    kw = dict(xycoords='data',textcoords="axes fraction",
              arrowprops=arrowprops, bbox=bbox_props, ha="right", va="top")
    ax.annotate(text, xy=(xmin, ymin), xytext=(0.55,0.9), **kw)

annot_min(df.X_k, df.fd_x)

plt.show()
```
